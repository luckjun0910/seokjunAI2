레모네이드 모델 만들기

```python
import pandas as pd
```

```python
import tensorflow as tf
```

```python
#파일들로 부터 데이터 읽어오기
파일경로 = 'https://raw.githubusercontent.com/luckjun0910/seokjunAI2/main/lemonade%20(4).csv'
레모네이드 = pd.read_csv(파일경로)
```

```python
#변수안에 어떤데이터가 있는지 확인(데이터 모양으로 확인)

print(레모네이드.shape)
```

```python
#칼럼 이름 출력
print(레모네이드.columns)
```

```python
독립 = 레모네이드[['온도']]
종속= 레모네이드[['판매량']]
print(독립.shape, 종속.shape)
#원인이 6개에 한줄 결과가 6개의 한줄
```

```python
#실제 데이터 출력

레모네이드.head()
```

```python
X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```python
model.fit(독립,종속,epochs=2000)
```

```python
print("predictions:",model.predict([20]))
```
